### 2.1 Visualisation of night-time site density

```python
import geopandas as gpd
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# 1. Data loading:
# Loading MSOA boundary data and nighttime stops data
msoa_boundary = gpd.read_file("MSOA_2011_London_gen_MHW.shp")
nighttime_stops_msoa = pd.read_csv("MSOA_nighttime_stops.csv")
```

```python
# 2. Data merging:
# Merging the two datasets based on the MSOA identifier
merged_data = msoa_boundary.merge(nighttime_stops_msoa, on='MSOA11CD', how='left')
```

```python
# Handle NaN values for stop_density
merged_data['NUMPOINTS'].fillna(0, inplace=True)
```

```python
# 3. Data visualisation:
# Plotting
fig, ax = plt.subplots(1, 1, figsize=(15, 10))
merged_data.plot(column='NUMPOINTS', cmap='Reds', legend=True, ax=ax)
ax.axis('off')
ax.set_title("Nighttime Stops Density in London MSOAs")
```
