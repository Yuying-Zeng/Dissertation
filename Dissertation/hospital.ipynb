```python
import geopandas as gpd
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# 1. 读取MSOA shapefile
msoa_gdf = gpd.read_file("MSOA_2011_London_gen_MHW.shp")
```

```python
# 2. 读取traffic sum CSV文件
workplace_df = pd.read_csv("hospital_sum.csv")
```

```python
# 3. 使用"category"列合并两个数据集
merged_gdf = msoa_gdf.merge(workplace_df, left_on='MSOA11CD', right_on='category', how='left')
```

```python
# 4. 使用geopandas可视化合并后的数据
fig, ax = plt.subplots(figsize=(15, 15))
merged_gdf.plot(column='mean', ax=ax, legend=True, cmap='viridis', edgecolor='black', 
                missing_kwds={"color": "lightgrey", "label": "Missing values"})
ax.set_title('Mean Accessibility by MSOA (Hospitals)')
ax.axis('off') 
plt.show()
```
